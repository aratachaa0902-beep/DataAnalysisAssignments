### Figure 2

Use `data-fig2.npy` for this figure.

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.load('data-fig2.npy')


x_data = a[:, 0]

y_red = a[:, 1]
y_green = a[:, 2]
y_blue = a[:, 3]

fig, ax = plt.subplots(figsize=(8, 5))


ax.plot(x_data, y_red,   color='red',   label='Red')
ax.plot(x_data, y_green, color='green', label='Green')
ax.plot(x_data, y_blue,  color='blue',  label='Blue')


ax.set_xlabel('X value')
ax.set_ylabel('Y value')

ax.legend()

plt.show()
```
