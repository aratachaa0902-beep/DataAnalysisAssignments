### Figure 2

Use `data-fig2.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt


data = np.load('data-fig2.npy')



x = np.arange(data.shape[0]) 
y_red = data[:, 0]
y_green = data[:, 1]
y_blue = data[:, 2]


fig, ax = plt.subplots()

ax.plot(x, y_red, color='red', label='Red')
ax.plot(x, y_green, color='green', label='Green')
ax.plot(x, y_blue, color='blue', label='Blue')

ax.set_xlabel('X value')
ax.set_ylabel('Y value')


ax.legend()

plt.show()

```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt


data = np.load('data-fig3.npy')


x = data[:, 0]
y = data[:, 1]

fig, ax = plt.subplots()


ax.scatter(x, y, s=200, color='blue', alpha=0.4)


ax.set_xlabel('X value')
ax.set_ylabel('Y value')

plt.show()

```

### Figure 4

Use `data-fig4.npy` for this figure.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


a = np.load('data-fig4.npy', allow_pickle=True)


print(a)


categories = ['A', 'B', 'C', 'D', 'E']
values = a 

fig, ax = plt.subplots()
ax.bar(categories, values, color='gray', alpha=0.7)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
plt.show()

```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

data = np.load('data-fig5.npy')


fig, ax = plt.subplots(figsize=(6, 6))


im = ax.imshow(data, cmap='RdBu', origin='lower', aspect='auto', vmin=-1, vmax=1)


fig.colorbar(im, ax=ax)


ax.set_xlabel('')
ax.set_ylabel('')

plt.show()

```
